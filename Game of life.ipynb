### Juego de la vida

* Emplear el programa del “juego de la vida” de John Conway para realizar el siguiente proceso de simulación:
    
    * Segun la circularidad de las vocales ver cuantas celulan quedan con vivas.
    
    * Determinar las variables que rigen el sistema
        * La variable a tomar en cuenta será la circularidad que tiene una letra, es decir la forma.
        
        La forma de las letra tienen curvas y lineas rectas, es este caso nos enfocamos en las vocalescomo podemos apreciar en las imagenes: 
        <div align="center">
    
    LETRA A - a
    
    <img src="a.png" width="500" height="500">
    
    LETRA E -e
    
        <img src="e.png" width="500" height="500"/>
    
    LETRA I - i
        
     <img src="i.png" width="500" height="500"/>
        
     Letra O - o
     
     <img src="o.png" width="500" height="500"/>
        
     Letra U - u
        
     <img src="u.png" width="500" height="500"/> </div>
        
    * Recopilar los resultados de estos planes. <br>
        Como podemos ver, las letras que tienen mas circularidad en su forma desaparecen todas la celulas como es el caso de la O mayúscula y la o minúscula, mientras que en vocales que tienen mas lineas rectas queda con una cantidad de celulas vivas.
